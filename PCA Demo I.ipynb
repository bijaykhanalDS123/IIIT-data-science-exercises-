## PCA Demonstration in Python

```python
# Importing the required libraries
```

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
```

```python
data.head()
```

```python
y = data.pop("Species")
```

```python
data.head()
```

```python
data.describe()
```

Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = data.copy()
```

```python
X.tail()
```

```python
x = scaler.fit_transform(X)
```

```python
x.shape
```

```python
x
```

```python
x[:5,:5]
```

```python
x.mean()
```

### Applying PCA on the data

```python
from sklearn.decomposition import PCA
```

```python
?PCA
```

```python
pca = PCA(random_state=42)
```

```python
pca.fit(x)
```

```python
pca.components_
```

```python
len(pca.components_)
```

```python
pca.components_[1]
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_)
```

```python
var_cumu = np.cumsum(pca.explained_variance_ratio_)
```

```python
var_cumu
```

#### Making the scree plot

```python
plt.plot(range(1,len(var_cumu)+1), var_cumu)
```

PCA with 2 components

```python
pc2 = PCA(n_components=2, random_state=42)
```

```python
newdata = pc2.fit_transform(x)
```

```python
newdata.shape
```

Making a dataframe out of it for convenience



```python
df = pd.DataFrame(newdata, columns=["PC1", "PC2"])
df.head()
```

```python
df_final = pd.concat([df, y], axis=1)
```

```python
df_final.head()
```

Plotting the data

```python
import seaborn as sns
```

```python
sns.pairplot(data=df_final, x_vars=["PC1"], y_vars=["PC2"], hue = "iris", size=5)
```
