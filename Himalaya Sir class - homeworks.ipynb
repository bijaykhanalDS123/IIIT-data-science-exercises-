### Exercise 1: Rewrite your pay computation with time-and-a-half for over-time and create a function called computepay which takes two parameters (hours and rate).

```python

def computepay(hours, rate):    
       
    if hours <= 40:
        
        pay = hours * rate
        
    else:
         
        pay = (hours - 40) * (rate *1.5) + 40 * rate
    
    return pay

print(computepay(float(input("Enter hours: ")), float(input("Enter rate: "))))
```

### Exercise 1: Write a program which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

```python
"""
to understand the count and total logic visualize the var as a cup and total as jar 

"""
total = 0

count = 0

while True:

    var = input("Enter your value: ") 
    
    if var == "done": # for valid string input "done"

      break

    else:
      
      try: # for valid integer input
      
        num = float(var) # converting the input to flaot

      except: # for invalid string input other than "done"

        print("Error") # when converting to float,there is an error,so print "error"  

        continue # goto the next iteration/ ask for next input so continue is used

      total = total + num # in each iteration the integer/float input is added to the value in the variable total 

      count = count + 1 # in each iteration the count is increased by 1

average = total/count

print("total: ",total, "\nCount: ", count,"\naverage:", average)



    
```
